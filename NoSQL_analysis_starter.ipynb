## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# review the collections in our database
db.list_collection_names()
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question: 
* Use `count_documents` to display the number of documents contained in the result.
* Display the first document in the results using `pprint`.
* Convert the result to a Pandas DataFrame, print the number of rows in the DataFrame, and display the first 10 rows.

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}
result = list(establishments.find(query))
# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query), "Documents in the result")
print("----")

# Display the first document in the results 
pprint(result[0])
```

```python
# convert to DataFrame
hygiene_df = pd.json_normalize(result)
print(f"{hygiene_df.shape[0]} rows in DataFrame")
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {"LocalAuthorityName": {"$regex": "London"}, "RatingValue": {"$gte": 4}}
result_2 = list(establishments.find(query_2))

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query_2), "Documents in the result")
print("----")

# Display the first document in the results 
pprint(result_2[0])
```

```python
# convert to DataFrame
rating_df = pd.json_normalize(result_2)
print(f"{rating_df.shape[0]} rows in DataFrame")
rating_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {"RatingValue": 5,
         "geocode.latitude": {"$gt": latitude - degree_search,
                              "$lt": latitude + degree_search},
         "geocode.longitude": {"$gt": longitude - degree_search,
                               "$lt": longitude + degree_search}}
sort = [("scores.Hygiene", 1)]

result_3 = list(establishments.find(query, sort=sort, limit=5))

pprint(result_3)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.DataFrame(result_3)
top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match = {"$match": {"scores.Hygiene": 0}}
group = {"$group": {"_id": "$LocalAuthorityName",
                    "count": {"$sum": 1}}}
sort = {"$sort": {"count": -1}}

pipeline = [match, group, sort]
result_4 = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(len(result_4), "Documents in results")

# Print the first 10 results
pprint(result_4[0:10])
```

```python
# Convert the result to a Pandas DataFrame
grouped_0_hygiene_df = pd.json_normalize(result_4)
print(f"{grouped_0_hygiene_df.shape[0]} rows in DataFrame")
grouped_0_hygiene_df.head(10)
```
